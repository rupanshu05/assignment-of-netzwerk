```python
# one hot encoding to convert categorical values to numerical value
df['Airline'].value_counts()
```

```python
Airline = df[['Airline']]
Airline = pd.get_dummies(Airline,drop_first=True)
Airline.head()
```

```python
df['Source'].value_counts()
```

```python
Source = df[['Source']]
Source = pd.get_dummies(Source,drop_first=True)
Source.head()
```

```python
df['Destination'].value_counts()
```

```python
Destination = df[['Destination']]
Destination = pd.get_dummies(Destination,drop_first=True)
Destination.head()
```

```python
df.replace({'non-stop' :0,
           '1 stop' :1,
           '2 stops' :2,
           '3 stops' :3,
           '4 stops' :4
           },inplace= True)
```

```python
df.head()
```

```python
#  Concatenate dataframe
df_1 = pd.concat([df, Airline, Source, Destination],axis=1)
df_1.head()
```

```python
df_1.drop(columns=['Airline','Source','Destination','Additional_Info','Route'],inplace=True)
df_1.head()
```

```python
df_1["Journey date"]=pd.to_datetime(df_1.Date_of_Journey,format="%d/%m/%Y").dt.day
df_1["Journey month"]=pd.to_datetime(df_1.Date_of_Journey,format="%d/%m/%Y").dt.month
```
