```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Car_sales.csv")
```

```python
df.drop('Latest_Launch',axis = 1,inplace= True)
```

```python
df.head()
```

```python
df = df.fillna(value = df['Price_in_thousands'].mean())
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head(50)
```

```python
df["Manufacturer"] = df["Manufacturer"].astype("category")
```

```python
df.dtypes
```

```python
df["Manufacturer"] = df["Manufacturer"].cat.codes
df.head()
```

```python
df["Model"] = df["Model"].astype("category")
df["Vehicle_type"] = df["Vehicle_type"].astype("category")
```

```python
df["Model"] = df["Model"].cat.codes
df["Vehicle_type"] = df["Vehicle_type"].cat.codes
df.head()
```

```python
x = df.loc[:, ['Manufacturer', "Model", "Sales_in_thousands", "__year_resale_value", "Vehicle_type", "Engine_size", "Horsepower","Wheelbase","Width","Length","Curb_weight","Fuel_capacity","Fuel_efficiency","Power_perf_factor"]]
```

```python
x
```

```python
y = df.loc[:,['Price_in_thousands']]
```

```python
y
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 10)
```

```python
model = LinearRegression()
```

```python
df.info()
```

```python
model.fit(x_train,y_train)
```

```python
df
```

```python
y_prediction = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_prediction)
```
