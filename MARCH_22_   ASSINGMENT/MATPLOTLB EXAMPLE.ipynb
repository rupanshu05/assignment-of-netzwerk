```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# scatter plot
```

```python
x = np.arange(0,20)
y = np.arange(30,50)
```

```python
plt.scatter(x,y, c= "g", marker = '*')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('scatter plot using matplotlib')

plt.show()
```

```python
# line plot

plt.plot(x,y,)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('line graph using matplotlib')

plt.show()
```

```python
df = [20,15,25,30,10,100]
```

```python
plt.boxplot(df)
plt.show()
```

```python
# class
x = ['8th','9th','10th']
# strength
y = [45,50,49]
```

```python
plt.bar(x,y,)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('bar graph using matplotlib')

plt.show()
```

```python
#pie chart
```

```python
x = [1000, 200, 300, 50]
labels =['china', 'US', 'russia', 'india']
colors = ['r', 'g', 'b', 'y']
```

```python
plt.pie(x, labels = labels, colors=colors , shadow = True)
plt.title("THis is as PIE Chart")
plt.show()
```

```python
e = (0.1,0,0,0)
plt.pie(x,explode = e, labels = labels, colors=colors , shadow = True)
plt.title("THis is as PIE Chart")
plt.show()
```

```python
e = (0.1,0,0,0)
plt.pie(x,explode = e, labels = labels, colors=colors ,autopct = "%1.2f%%" , shadow = True)
plt.title("THis is as PIE Chart")
plt.show()
```
